Mental health risk is not caused by one factor alone. Users with high screen time, reduced sleep and late-night usage show the highest depression and anxiety. This demonstrates a compounding behavioral risk effect.

```python
display(spark.sql("""
SELECT *
FROM ecommerce.mental_health.gold_behavioral_risk_patterns
ORDER BY Risk_Percentage DESC
LIMIT 10
"""))
```

Exposure to comparison-driven content significantly increases anxiety levels, highlighting the psychological cost of curated social feeds.

```python
display(spark.sql("""
SELECT 
    Social_Comparison_Risk,
    AVG(High_Mental_Health_Risk) AS Risk_Rate
FROM ecommerce.mental_health.silver_user_features
GROUP BY Social_Comparison_Risk
"""))
```

High screen time users show elevated mental distress, supporting research linking digital overexposure to emotional fatigue.

```python
display(spark.sql("""
SELECT 
    High_Screen_Time,
    AVG(High_Mental_Health_Risk) AS Risk_Rate
FROM ecommerce.mental_health.silver_user_features
GROUP BY High_Screen_Time
"""))
```

Users active after midnight sleep less and show higher depression scores, indicating circadian disruption as a key digital health risk.

```python
display(spark.sql("""
SELECT 
    LateNight_Sleep_Risk,
    AVG(High_Mental_Health_Risk) AS Risk_Rate
FROM ecommerce.mental_health.silver_user_features
GROUP BY LateNight_Sleep_Risk
"""))
```
